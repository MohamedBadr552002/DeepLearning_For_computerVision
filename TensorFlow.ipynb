## Linear Algebra

```python
# Matrix Multiplication
# tf.matmul(x,y) == (x@y)
print(tf.matmul(tf.constant([1, 2, 3, 4, 5, 6], shape=[2, 3]) , tf.constant([7, 8, 9, 10, 11, 12], shape=[3, 2]) ))
```

```python
arr_1 = tf.constant([1, 2, 3, 4, 5, 6], shape=[2, 3])
print(tf.transpose(arr_1))
```

```python
x_1 = tf.constant([
    [ [9,3],
     [10,5]],

    [ [9,3],
     [10,5]],

    [ [9,3],
     [10,5]]
    ])
'''
for 3D transpose turn any element [i, j, k] to [k, j, i]
'''
print(x_1[0,1,1])
x_2 = tf.transpose(x_1)
print(x_2)
```

```python
x_3 = tf.constant([
    [ [9,3,7],
     [10,5,7]],

    [ [9,3,7],
     [10,5,7]],

     [ [9,3,7],
     [10,5,7]]
    ])

print(x_1.shape)
print(x_3.shape)

'''
For Right Matrix Multiplication the size of the matrices should (i, j, k) and (i, k, j)

In 3D matrix Multiplication mutiple each element of matrix in the veiw of it using matrix multiplication
'''
print(tf.matmul(x_1,x_3))
```

## TenserFlow Special Function

**tf.expand_dims**

```python
input = tf.constant([1, 2, 3, 4, 5, 6], shape=[2, 3])
print(input)
print(input.shape)
print("----------------------------------")

expanded_output = tf.expand_dims(input, 0, name=None)
print(expanded_output)
print(expanded_output.shape)
```

**squeeze**

```python
#squeezed_out = tf.squeeze(input, axis=0, name=None ) #ERROR -> matrix should be expanded to be squeezed
#squeezed_out = tf.squeeze(expanded_output, axis=1, name=None ) #ERROR -> Must be the same axis used in expanasion
squeezed_out = tf.squeeze(expanded_output, axis=0, name=None )
Tensor_data(expanded_output)
Tensor_data(squeezed_out)
```

**Reshape**

```python
# -1 used to automatically reshap
reshaped_out = tf.reshape(input,[-1] , name=None)
Tensor_data(input)
Tensor_data(reshaped_out)
reshaped_out = tf.reshape(input , [2, 1, 3])
Tensor_data(reshaped_out)
reshaped_out = tf.reshape(input , [3, 1,-1])
Tensor_data(reshaped_out)
```
